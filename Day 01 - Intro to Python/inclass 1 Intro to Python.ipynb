```python
var1 = 1
print(var1)
print("address of var1 is: ", id(var1)) # unique identifier of the object that var1 is pointing to
print("Hex address of var1 is: ", hex(id(var1)))
print(type(var1))

var2 = 3.14 # float
print(var2)
print(type(var2))

var3 = "idrees" # string
print(var3)
print(type(var3))

var1 = 5 # reassignment of a variable
print(var1)


var5 = None
print(var5)
print(type(var5))
```

# Operations

- how we operate or process different variables

- e.g.
  - addition: `+`
  - sub, mul, div

- operations are valid only between similar data types

```python
"idrees" + "idrees" # concatenation ---> joining back to back

10 * "idrees"

10 * "--"
```

```python
var1 = 10
var2 = 3

# arthematic operators
# +, -, *, /, **, %, //,

print(var1 + var2)

print(var1 - var2)

print(var2 - var1)

print(var1 * var2)

print(var1 / var2)

print(var1 // var2) # floor division

print(var1 % var2) # modulus operation ---> modular division

print(var1 ** var2) # exponent ---> Raised to power
```

```python
# comparison operators

# these help us compare two objects: ==, !=, <, >, <=, >=
# these result in boolean values

var1 = 10 # assignment operator ---> helps assign values to variables
var2 = 3

var1 == var2 # result of comparision operations are boolean values

var1 < var2

var1 >= var2
```

```python
# logical operators

# used to combine conditional statements

(10 < 0) or (0 > 2)

False or False

# think of Or as addition
False or True

# to remember think of AND as multiplication
True and True

# not inverts
# it will take just one input
not (10 < 0)
```

```python
# in comes in when you learn about sequences
"i" in "idrees" 


"x" in "idrees"
```

```python
# chained comparisons

# set x between a range: we say this:  start < x < end

age = 18 

18 < age < 100


# 18 < 18 ---> False ---because < is EXCLUSIVE
# 18 <= 18 ---> True ---because <= is INCLUSIVE
```
